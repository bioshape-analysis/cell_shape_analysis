```python
### Run this first to calculate all needed quantities for teh next plots (which are boxplot of angles and two regeressions)
import pickle
import numpy as np
import matplotlib.pyplot as plt
import pandas

### load data
with open("data/W_all.pkl", "rb") as file:
    W = pickle.load(file)
with open("data/C_all.pkl", "rb") as file:
    C = pickle.load(file)
with open("data/T_all.pkl", "rb") as file:
    T = pickle.load(file)

### functions
def angle_between_vectors(v1, v2):
    """
    Calculates the angle between two vectors using the dot product.

    Args:
        v1 (np.ndarray): The first vector.
        v2 (np.ndarray): The second vector.

    Returns:
        float: The angle in radians (range [0, pi]).
    """
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    if norm_v1 == 0 or norm_v2 == 0:
        return 0.0
    cosine_angle = np.clip(dot_product / (norm_v1 * norm_v2), -1.0, 1.0)
    angle_rad = np.arccos(cosine_angle)
    return angle_rad


phys_turn=list()
phys_dist=list()
shape_change=list()
Disp_list=list()
Wmax_list=list()
dpratio_phys=list()
dpratio_shape=list()
for i in range(len(W)):
  #if i in ex:
    #continue
  sorted_Centroid=C[i]
  v=C[i]
  Wmat2=W[i]
  sorted_Time=T[i]
  ### calculate velocity vectors for physical movement and angles between them
  velocity_vectors = np.diff(sorted_Centroid, axis=0)
  angles = []
  for j in range(len(velocity_vectors) - 1):
      v1 = velocity_vectors[j]
      v2 = velocity_vectors[j+1]
      angle_rad = angle_between_vectors(v1, v2)
      angles.append(np.degrees(angle_rad))
  ### extract distances between shapes
  sup_diag = np.diag(Wmat2, k=1)

  ### Centroid curve
  from scipy import stats
  dv=[]
  dist_inc=0
  for k in range(1,v.shape[0]):
    inc=v[k,:]-v[k-1,:]
    dist_inc=dist_inc+np.linalg.norm(inc,2)
    dv.append(dist_inc)
  DCent=dv[-1]#np.array(dv)

  an=np.array(angles)
  sh=np.array(sup_diag)
  N=sorted_Time[-1]-sorted_Time[0]

  ### "straight line" displacement for the path
  Disp_list.append(np.linalg.norm(sorted_Centroid[-1,:]-sorted_Centroid[0,:])/N)

  ### maximal Wasserstein distance
  Wmax_list.append(np.max(Wmat2)/N)

  ### displacement over pathlength
  Disp_phys=np.linalg.norm(sorted_Centroid[-1,:]-sorted_Centroid[0,:])
  Pathl_phys=DCent
  dpratio_phys.append(Disp_phys/Pathl_phys)

  Disp_shape=Wmat2[0,-1]
  Pathl_shape=np.sum(sh)
  dpratio_shape.append(Disp_shape/Pathl_shape)

  ### calculate average quantities
  phys_turn.append(np.sum(an)/N)
  phys_dist.append(DCent/N)
  #shape_change.append(np.sum(np.square(sh))/N)
  shape_change.append(np.sum(sh)/N)

p=np.array(phys_turn)
vbarphys=np.array(phys_dist)
vbarshape=np.array(shape_change)
dis=np.array(Disp_list)
wmax=np.array(Wmax_list)
rphys=np.array(dpratio_phys)
rshape=np.array(dpratio_shape)



```

```python
################################ plot angles for straight and non-straight (boxplot) and conduct two-sample t-test
pturn=np.array(phys_turn)

qlow=np.quantile(rphys,0.15)
cell_indices_low=np.where(np.array(rphys<qlow))[0]

qhigh=np.quantile(rphys,0.85)
cell_indices_high=np.where(np.array(rphys>qhigh))[0]

pstraight=[pturn[i] for i in cell_indices_high]
pchaotic=[pturn[i] for i in cell_indices_low]


### data on angle
d_plot = [pstraight,pchaotic]

fig, ax = plt.subplots()
ax.boxplot(d_plot, medianprops=dict(color='gray', linewidth=2), labels=['straight', 'non-straight'])
ax.tick_params(axis='both', labelsize=18)
ax.set_ylabel('angle', fontsize=20)
ax.set_xlabel('')
plt.savefig("fig3_boxplot.svg")
### conduct two-sample t test for the difference between two groups
from scipy import stats
t_statistic, p_value = stats.ttest_ind(np.array(pstraight), np.array(pchaotic))

print(f"p_value for two sample t test is {p_value}")

# ########################################### uncomment this if you want colors
# plt.style.use('ggplot')
# colors = ['deepskyblue', 'lime']
# bp = plt.boxplot(d_plot, patch_artist=True, labels=['straight', 'non-straight'], medianprops=dict(color='gray', linewidth=2))
# for patch, color in zip(bp['boxes'], colors):
#     patch.set_facecolor(color)
#     #ax.tick_params(axis='x', labelbottom=False, labelleft=False)
#     #ax.tick_params(axis='both', labelsize=18)
#     plt.gca().get_yaxis().set_visible(False)
#     #plt.gca().get_yaxis().set_visible(False)
#     #plt.yscale("log")
```

```python
####################################### Plot regression for vbar_phys/vbar_shape
from scipy import stats

# Define your data: vbarphys/vbarshape
x = np.array(vbarphys)
y = np.array(vbarshape)


slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Print the detailed results
print('regression results for vbar')
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r_value**2}")
print(f"P-value: {p_value}")
print(f"Standard error: {std_err}")

x_line = np.linspace(min(x), max(x), 100)
y_pred = slope * x_line + intercept

plt.style.use('default')  
### plot results for vbar
plt.figure(figsize=(8, 6))

# Plot data and regression line in gray color
plt.scatter(x, y, label='vbar data', color='black',s=30)
plt.plot(x_line, y_pred, label='Regression Line', color='black')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('vbar phys')
plt.ylabel('vbar shape')
plt.savefig("fig4a.svg")
plt.show()
```

```python
x = np.array(rphys)
y = np.array(rshape)

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

x_line = np.linspace(min(x), max(x), 100)
y_pred = slope * x_line + intercept
plt.style.use('default')  
plt.figure(figsize=(8, 6))

plt.scatter(x, y, label='SI data Data', color='black',s=70)
plt.plot(x_line, y_pred, label='Regression Line', color='black')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('SI phys')
plt.ylabel('SI shape')
plt.show()

plt.style.use('default')  
plt.figure(figsize=(8, 6))

qlow=np.quantile(rphys,0.15)
cell_indices_low=np.where(np.array(rphys<qlow))[0]

qhigh=np.quantile(rphys,0.85)
cell_indices_high=np.where(np.array(rphys>qhigh))[0]


group=np.zeros((1,len(rphys)),dtype=int)
for i in range(len(rphys)):
  if i in cell_indices_high:
    group[0,i]=int(1)
  elif i in cell_indices_low:
    group[0,i]=int(2)

group=group.ravel().tolist()
group=[str(num) for num in group]

data = {'x': x,
        'y': y,
        'group': group}
df = pandas.DataFrame(data)
color_map = {'0': 'black', '1': 'deepskyblue', '2': 'lime'}
colors = [color_map[group] for group in df['group']]

plt.scatter(df['x'], df['y'], c=colors, s=30)
plt.plot(x_line, y_pred, label='Fitted Regression Line', color='gray')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.savefig("fig4b.svg")
plt.show()
```
